```python
from util_functions.aggregation_utils import aggregate_entities, named_entity_recognition, is_person_name
# testing code for presenter extraction
def extract_all_presenters(df, award):
    tweets = df[df['cleaned_text'].str.lower().str.contains('present')]['cleaned_text']

    def remove_punctuation(text):
        return ''.join(char for char in text if char.isalnum() or char.isspace())

    # filter tweets that contain the award name without punctuation
    tweets = tweets[tweets.apply(lambda x: remove_punctuation(award).lower() in remove_punctuation(x).lower())]

    return tweets.tolist()

presenters = extract_all_presenters(df, "best motion picture drama")

presenters_entities = named_entity_recognition(presenters)
# Filter out entities that aren't person names
presenters_entities = [entity for entity in presenters_entities if is_person_name(entity['Name'])]

print(presenters_entities)
```
